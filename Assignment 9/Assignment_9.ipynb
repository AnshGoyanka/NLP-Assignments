```python
from transformers import pipeline
```

```python
sentiment = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)
```

```python
text = "यह फिल्म बहुत अच्छी है"

print(sentiment(text))
```

```python
text = "हा चित्रपट खूप छान आहे"

print(sentiment(text))
```

```python
text = "இந்த படம் அருமையாக உள்ளது"

print(sentiment(text))
```

```python
def get_sentiment_label(text):

    result = sentiment(text)[0]['label']

    stars = int(result[0])

    if stars <= 2:
        return "Negative"
    elif stars == 3:
        return "Neutral"
    else:
        return "Positive"
```

```python
print(get_sentiment_label("यह फिल्म बहुत अच्छी है"))
```

```python
def sentiment_app():
    print("Indian Language Sentiment Analyzer (type exit to stop)")

    while True:
        text = input("Enter text: ")

        if text.lower() == "exit":
            break

        print("Sentiment:", get_sentiment_label(text))

sentiment_app()
```
