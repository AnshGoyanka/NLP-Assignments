```python
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
```

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn
```

```python
sentence = "I went to the bank to deposit money"

words = word_tokenize(sentence)

sense = lesk(words, "bank")

print("Sense:", sense)
print("Definition:", sense.definition())
```

```python
sentence = "The river bank was full of trees"

words = word_tokenize(sentence)

sense = lesk(words, "bank")

print("Sense:", sense)
print("Definition:", sense.definition())
```

```python
def wsd_word(sentence, target_word):

    words = word_tokenize(sentence)

    sense = lesk(words, target_word)

    if sense:
        print("Word:", target_word)
        print("Sense:", sense.name())
        print("Definition:", sense.definition())
        print("Example:", sense.examples())
    else:
        print("No sense found")
```

```python
wsd_word("I deposited money in the bank", "bank")

wsd_word("The river bank is very beautiful", "bank")
```

```python
for syn in wn.synsets("bank"):
    print(syn.name(), ":", syn.definition())
```
