```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from collections import Counter, defaultdict
```

```python
text = """
I love machine learning.
I love artificial intelligence.
I love natural language processing.
Machine learning is amazing.
Artificial intelligence is the future.
"""
```

```python
tokens = word_tokenize(text.lower())
```

```python
n = 2

ngrams_list = list(ngrams(tokens, n))
ngram_counts = Counter(ngrams_list)
```

```python
model = defaultdict(list)

for (w1, w2), count in ngram_counts.items():
    model[w1].append((w2, count))
```

```python
def autocomplete(word, top_k=3):

    if word not in model:
        return "No prediction found"

    predictions = sorted(
        model[word],
        key=lambda x: x[1],
        reverse=True
    )

    return [p[0] for p in predictions[:top_k]]
```

```python
print(autocomplete("love"))
```

```python
n = 3
trigrams_list = list(ngrams(tokens, n))
trigram_counts = Counter(trigrams_list)

tri_model = defaultdict(list)

for (w1, w2, w3), count in trigram_counts.items():
    tri_model[(w1, w2)].append((w3, count))
```

```python
def autocomplete_tri(w1, w2):

    if (w1, w2) not in tri_model:
        return "No prediction"

    predictions = sorted(
        tri_model[(w1, w2)],
        key=lambda x: x[1],
        reverse=True
    )

    return predictions[0][0]
```

```python
print(autocomplete_tri("i", "love"))
```
